```python
import pandas
from functools import reduce
```

```python
dataframes = [
    pandas.read_csv(_ + "_params.csv")
    for _ in ["esmif", "pythia", "foldx", "aa_baseline", "ddgun3d",'so']
]
#dataframes
```

```python
params_df = reduce(
    lambda left, right: pandas.merge(left, right, on="Unnamed: 0", how="outer"),
    dataframes,
).rename({"Unnamed: 0": "Parameter"}, axis=1)
params_df
```

```python
params_df.to_csv('ddmbc_params.csv',index=False)
```

```python
noscale_df=pandas.read_csv('ddmbc_params.csv',index_col='Parameter').drop('delta')
noscale_df
```
